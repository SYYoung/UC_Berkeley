```python
from datascience import *
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
data = "../Data/nc-est2015-agesex-res.csv"
full =Table.read_table(data)
full
```

```python
partial = full.select('SEX', 'AGE', 4, 9)
partial
```

```python
us_pop = partial.relabeled(2, '2010'). relabeled(3, '2015')
```

```python
us_pop
```

```python
us_pop.set_format([2,3], NumberFormatter)
```

```python
us_pop = us_pop.with_columns('Change', us_pop.column('2015')-us_pop.column('2010'))
```

```python
min(us_pop.column('Change'))
```

```python
us_pop.where('Change', are.equal_to(min(us_pop.column('Change'))))
```

```python
us_pop.where('AGE', are.equal_to(47-5))
```

```python
us_pop.where('AGE', are.equal_to(47+5))
```

```python
us_pop.sort('Change', descending=True)
```

```python
us_pop = us_pop.with_column('Total Growth', us_pop.column('Change')/us_pop.column('2010'))
us_pop.set_format('Total Growth', PercentFormatter)
```

```python
help(us_pop.set_format)
```

```python
us_pop.sort('Change', descending=True)
```

```python
us_pop_2015 = us_pop.drop('2010').where('AGE', are.below(999)).where('SEX', are.above(0))
us_pop_2015
```

```python
us_pop_2015 = us_pop_2015.drop('Change').drop('TotalGrowth')
us_pop_2015 = us_pop_2015.drop('Total Growth')
us_pop_2015
```

```python
males = us_pop_2015.where('SEX', 1).column('2015')
by_sex = us_pop_2015.where('SEX', 2).drop('SEX').relabeled('2015', 'Females').with_column('Males', males)
by_sex.set_format('Males', NumberFormatter)
```

```python
by_sex.plot('AGE')
```
